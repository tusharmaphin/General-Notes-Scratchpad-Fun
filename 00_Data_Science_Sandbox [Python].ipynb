```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the number of steps (time)
N = 1000

# 2. Simulate the quantum coin flips (-1 or 1)
# Think of this as the random forces hitting our particle
steps = np.random.choice([-1, 1], size=N)

# 3. Calculate the position over time
# We integrate (sum) the steps to get the final path
position = np.cumsum(steps)

# 4. Plot the trajectory
plt.figure(figsize=(10, 5))
plt.plot(position, color='crimson')
plt.title("1D Random Walk (Brownian Motion / Simulated Stock Price)")
plt.xlabel("Time (Steps)")
plt.ylabel("Position (Price)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

options = [-1, +1]
distribution = {}

def simulate_walk(steps):
  x_0 = 0
  for i in range(steps):
    x_0 += random.choice(options)
  return x_0

for i in range(10000):
  x = simulate_walk(100)
  k = 1
  if x in distribution:
    distribution[x] += 1
  else:
    distribution[x] = k

positions = list(distribution.keys())
frequencies = list(distribution.values())

plt.bar(positions, frequencies)
plt.show()
```

```python
#R_t = (P_t - P_{t-1})/P_{t-1}.

import numpy as np
Stock = np.array([100.0, 102.0, 99.0, 101.0, 105.0])
P_t = Stock[1:5]
P_tminus1= Stock[:4]

R_t = (P_t - P_tminus1)/P_tminus1
sigma = np.std(R_t) #Also known as volatility in Finance
print(R_t, sigma)
```

```python
import random
total_flips = 10
Trials = [10, 10000, 1000000]
Prob = {}

for t in Trials:
  head_counts = 0
  for i in range(t):
    x = random.choice([0, 1])
    if x == 1:
      head_counts += 1
    Prob[t] = head_counts/t

print(Prob)
```

```python
#Logistic Regression
import numpy as np

# Raw z-scores (linear outputs) for 5 users
z_scores = np.array([-5.0, -1.2, 0.0, 1.5, 4.0])

def sigmoid(z):
  sigma = 1/(1 + np.exp(-z))
  return sigma

print(sigmoid(z_scores))
```

```python
#Confusion-Matrix
import numpy as np

# 1 = Fraud, 0 = Legitimate
y_true = np.array([1, 0, 1, 1, 0, 1, 0, 0])
y_pred = np.array([1, 0, 0, 1, 0, 1, 1, 0])
def manual_confusion_matrix(y_true, y_pred):
  TP = np.sum((y_true == 1) & (y_pred == 1))
  TN = np.sum((y_true == 0) & (y_pred == 0))
  FP = np.sum((y_true == 0) & (y_pred == 1))
  FN = np.sum((y_true == 1) & (y_pred == 0))
  x = np.array([[TP, FP], [FN, TP]])/len(y_true)
  return x

print(manual_confusion_matrix(y_true, y_pred))
```

```python
#Decision Trees
import numpy as np

node_A = np.array([1, 1, 1, 1, 1])       # Pure state
node_B = np.array([1, 0, 1, 0, 1, 0])    # High entropy state
node_C = np.array([1, 1, 1, 0])          # Mixed state

def calculate_gini(node):
  k = len(node)
  p1 = np.mean(node) #The Fraction of 1s
  p0 = 1 - p1
  G = 1 - p1**2 - p0**2
  return G

print(calculate_gini(node_A), calculate_gini(node_B), calculate_gini(node_C))
```
