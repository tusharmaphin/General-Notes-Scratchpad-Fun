```
import pandas as pd

import numpy as np



df = pd.read_csv('https://raw.githubusercontent.com/tusharmaphin/General-Notes-Scratchpad-Fun/refs/heads/main/south_park_ratings.csv')



df['Weight'] = np.where(df['Rating'] >= 7.2, 1.0, np.exp(1.5*(df['rating'] - 7.2)))

df['Normalized_Weight'] = df.['Weight']/(df['Weight'].sum())



df.sample(n = 1, df['Normalized_Weight'])

```
