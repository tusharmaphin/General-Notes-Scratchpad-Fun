```
#South Park (2017)
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/tusharmaphin/General-Notes-Scratchpad-Fun/refs/heads/main/data/south_park_ratings.csv')

df['Weight'] = np.where(df['Rating'] >= 7.2, 1.0, np.exp(1.5*(df['Rating'] - 7.2)))
df['Normalized_Weight'] = df['Weight']/(df['Weight'].sum())

recommendation = df.sample(n = 1, weights = 'Normalized_Weight')
title = recommendation['Title'].values[0]
rating = recommendation['Rating'].values[0]

print(f"Tonight's Recommendation: {title} (Rating: {rating})")
```

```
#The Office (2005)
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/tusharmaphin/General-Notes-Scratchpad-Fun/refs/heads/main/data/the_office_ratings.csv')

df['Weight'] = np.where(df['Rating'] >= 7.5, 1.0, np.exp(1.0*(df['Rating'] - 7.2))) #Rate changed from 1.5 to 1.0 to make it less sloppy.
df['Normalized_Weight'] = df['Weight']/(df['Weight'].sum())

recommendation = df.sample(n = 1, weights = 'Normalized_Weight')
title = recommendation['Title'].values[0]
rating = recommendation['Rating'].values[0]

print(f"Tonight's Recommendation: {title} (Rating: {rating})")
```
